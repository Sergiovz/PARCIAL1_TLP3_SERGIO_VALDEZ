### 3. Resolución

- Manejo de valores nulos: "Precio_Unitario", "Total_Venta",

- Transformamos la columna "Fecha" con tipo de dato (object) de nuestro DataFrame al tipo datetime utilizando "pd.to_datetime()", primero transformamos nuestra columna fecha en una Series con "pd.Series()", luego creamos dos variables nuevas y en cada variable transformamos al tipo de fecha con su formato correspondiente, por ultimo combinamos ambas Series con "combine_firts()"

- Normalizamos los nombres de las columnas "Nombre_Cliente" y "Apellido_Cliente" Capitalizandolos con str.title(), luego mostramos el DataFrame con los primeros valores, dandonos cuenta se normalizo con exito (por ejemplo: sofía martínez).

```python

df["Precio_Unitario"] = df["Precio_Unitario"].fillna(df["Precio_Unitario"].median())
df["Total_Venta"] = df["Total_Venta"].fillna(df["Total_Venta"].median())
```

```python
fecha_series = pd.Series(df["Fecha"])

fecha_series1 = pd.to_datetime(fecha_series, format="%Y-%m-%d", errors="coerce")
fecha_series2 = pd.to_datetime(fecha_series, format="%d-%m-%Y", errors="coerce")

fecha_series_transformada = fecha_series1.combine_first(fecha_series2)
fecha_series_transformada

df["Fecha"] = fecha_series_transformada
```

```python
df["Nombre_Cliente"] = df["Nombre_Cliente"].str.title()
df["Apellido_Cliente"] = df["Apellido_Cliente"].str.title()

df.head()
```

- Exportamos el archivo (dataframe) a un archivo SQLite, luego realizamos una conexión y haccemos una consulta para verificar que todo haya salido bien
